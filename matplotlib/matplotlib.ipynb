```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#simple plot( a basic ghraph)
y_points = np.array([3,8,1,10,5,7])
plt.plot(y_points)
plt.show()
```

```python
#a line
x_points = np.array([1,8])
y_points = np.array([3,10])
plt.plot(x_points,y_points)
plt.show()
```

```python
#ploting without line
x_points = np.array([1,8])
y_points = np.array([3,10])
plt.plot(x_points,y_points,'o') #o is for dots
plt.show()
```

```python
#showing the table of marker options
x_points = np.array([1,8])
y_points = np.array([3,10])
plt.plot(x_points,y_points,marker = '3') #marker = '3' is a triangle
plt.show()
```

```python
#changing the color of the line
x_points = np.array([1,8])
y_points = np.array([3,10])
plt.plot(x_points,y_points,color = 'r') #color = 'r' is red
plt.show()
```

```python
#set multiple options
y_points = np.array([3,8,1,10,5,7])
plt.plot(y_points, marker = 'o', ms = 20, mec = 'r', mfc = 'r', linestyle = 'dotted') #ms is marker size, mec is marker edge color, mfc is marker face color
plt.show()
```

```python
plt.plot(y_points,'o-g') #o is for dots, - is for line, g is for green
plt.show()
```

```python
#plotting multiple lines
y_points = np.array([3,8,1,10,5,7])
y_points2 = np.array([6,2,7,11,4,8])
plt.plot(y_points)
plt.plot(y_points2)
plt.show()
```

```python
#plotting multiple lines with labels
x_points = np.array([1,2,3,4])
y_points = np.array([1,4,9,16])

plt.plot(x_points,y_points)
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.xlabel('X-axis',fontdict = font1)
plt.ylabel('Y-axis',fontdict = font2)
plt.title('Simple plot',fontdict = font1)
plt.show()
```

```python
#Pie Chart
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0, 0.5, 0, 0.1]

plt.pie(y, labels = mylabels , explode = myexplode)
plt.show() 
```

```python
x = [0,1,2,3,4]
y = [0,2,4,6,8]

# Resize your Graph (dpi specifies pixels per inch. When saving probably should use 300 if possible)
plt.figure(figsize=(8,5), dpi=100)

# Line 1

# Keyword Argument Notation
#plt.plot(x,y, label='2x', color='red', linewidth=2, marker='.', linestyle='--', markersize=10, markeredgecolor='blue')

# Shorthand notation
# fmt = '[color][marker][line]'
plt.plot(x,y, 'b^--', label='2x')

## Line 2

# select interval we want to plot points at
x2 = np.arange(0,4.5,0.5)

# Plot part of the graph as line
plt.plot(x2[:6], x2[:6]**2, 'r', label='X^2')

# Plot remainder of graph as a dot
plt.plot(x2[5:], x2[5:]**2, 'r--')

# Add a title (specify font parameters with fontdict)
plt.title('Our First Graph!', fontdict={'fontname': 'Comic Sans MS', 'fontsize': 20})

# X and Y labels
plt.xlabel('X Axis')
plt.ylabel('Y Axis')

# X, Y axis Tickmarks (scale of your graph)
plt.xticks([0,1,2,3,4,])
#plt.yticks([0,2,4,6,8,10])

# Add a legend
plt.legend()

# Save figure (dpi 300 is good when saving so graph has high resolution)
plt.savefig('mygraph.png', dpi=300)

# Show plot
plt.show()
```
