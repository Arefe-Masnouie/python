```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#simple plot( a basic ghraph)
y_points = np.array([3,8,1,10,5,7])
plt.plot(y_points)
plt.show()
```

```python
#a line
x_points = np.array([1,8])
y_points = np.array([3,10])
plt.plot(x_points,y_points)
plt.show()
```

```python
#ploting without line
x_points = np.array([1,8])
y_points = np.array([3,10])
plt.plot(x_points,y_points,'o') #o is for dots
plt.show()
```

```python
#showing the table of marker options
x_points = np.array([1,8])
y_points = np.array([3,10])
plt.plot(x_points,y_points,marker = '3') #marker = '3' is a triangle
plt.show()
```

```python
#changing the color of the line
x_points = np.array([1,8])
y_points = np.array([3,10])
plt.plot(x_points,y_points,color = 'r') #color = 'r' is red
plt.show()
```

```python
#set multiple options
y_points = np.array([3,8,1,10,5,7])
plt.plot(y_points, marker = 'o', ms = 20, mec = 'r', mfc = 'r', linestyle = 'dotted') #ms is marker size, mec is marker edge color, mfc is marker face color
plt.show()
```

```python
plt.plot(y_points,'o-g') #o is for dots, - is for line, g is for green
plt.show()
```

```python
#plotting multiple lines
y_points = np.array([3,8,1,10,5,7])
y_points2 = np.array([6,2,7,11,4,8])
plt.plot(y_points)
plt.plot(y_points2)
plt.show()
```

```python
#plotting multiple lines with labels
x_points = np.array([1,2,3,4])
y_points = np.array([1,4,9,16])

plt.plot(x_points,y_points)
font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.xlabel('X-axis',fontdict = font1)
plt.ylabel('Y-axis',fontdict = font2)
plt.title('Simple plot',fontdict = font1)
plt.show()
```

```python
#plotting multiple lines with labels
x_points = np.array([1,2,3,4])
y_points = np.array([1,4,9,16])

plt.title("simple graph")
plt.xlabel("x_axis")
plt.ylabel("y_axis")

plt.plot(x_points,y_points)

plt.grid(color = 'green', linestyle = '--', linewidth = 0.5) #grid color is green, line style is dashed, line width is 0.5
plt.show()
```

```python
#subplot
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 1)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 2)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 3)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 4)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 3, 5)
plt.plot(x,y)

x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 3, 6)
plt.plot(x,y)

plt.show()
```
